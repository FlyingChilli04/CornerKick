```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["binary_result"] = (matches["result"] == 'W').astype("int")
```

```python
matches
```

```python
# Sort all teams in the data by date
matches.sort_values(by='date', inplace=True)
```

```python
# Sort all data by teams
matches.sort_values(by='team', inplace=True)
```

```python
matches = matches.groupby('team').apply(lambda x: x.sort_values(by='date'))
```

```python
matches.reset_index(drop=True, inplace=True) # check later
```

```python
matches['prev_possession'] = matches.groupby('team')['poss_y'].shift(1)
```

```python
matches['prev_shots_on_target'] = matches.groupby('team')['sh'].shift(1)
```

```python
matches = matches.groupby('team').apply(lambda x: x.iloc[1:]).reset_index(drop=True)
```

```python
matches
```

```python
matches["target"] = matches["ck"]
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# n_estimators, number of trees in the forest
# min_samples_split, min number of samples in a leaf of a dec tree (higher, less overfitting, but lower accuracy)
# random_state, get same results as long data the same
rf = RandomForestRegressor(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
# split data into training and test data
train = matches[matches["date"] < '2022-01-01']
test = matches[matches["date"] > '2022-01-01']
```

```python
# list of predictors
predictors = ["venue_code", "opp_code", "prev_shots_on_target", "prev_possession"]
```

```python
# train random forest model
rf.fit(train[predictors], train["target"])
```

```python
pred = rf.predict(test[predictors])
```

```python
# get mean absolute error of our predictions to targets
from sklearn.metrics import mean_absolute_error
```

```python
error = mean_absolute_error(test["target"], pred)
```

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Arsenal")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "xg", "xga", "poss_x", "ck", "touches", "binary_result"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
# make unique indices
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
# retrain model with new predictors (rolling averages)
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    pred = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=pred), index=test.index)
    error = mean_absolute_error(test["target"], pred)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```
